# SeamGPT Mesh Processing Pipeline

This notebook provides an interactive way to run the mesh processing pipeline.


```python
from seamgpt_mesh_pipeline import MeshProcessor

# Configure parameters
input_dir = "meshes"
output_dir = "outputs"
bins = 1024

# Create processor and run
processor = MeshProcessor(input_dir, output_dir, bins)
processor.run()
```

## View Results

After running the pipeline, you can inspect the results:


```python
import json
from pathlib import Path

# Load metrics
metrics_path = processor.run_output_dir / "metrics.json"
with open(metrics_path) as f:
    metrics = json.load(f)

print(json.dumps(metrics, indent=2))
```

```python
# Display report
report_path = processor.run_output_dir / "REPORT.md"
with open(report_path) as f:
    print(f.read())
```
